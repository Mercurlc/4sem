```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and axis to plot on
fig, ax = plt.subplots(figsize=(8, 8))

# Variables
x = np.linspace(0, 10, 400)
y1 = (7 - x)/2
y2 = 8 - 2*x
y3 = np.full_like(x, 3)

# Plot the lines
ax.plot(x, y1, label=r'$x_1 + 2x_2 \leq 7$')
ax.plot(x, y2, label=r'$2x_1 + x_2 \leq 8$')
ax.hlines(3, 0, 10, label=r'$x_2 \leq 3$', colors='orange')

# Fill feasible region
y4 = np.minimum(y1, y3)
y5 = np.minimum(y4, y2)
ax.fill_between(x, np.maximum(0, y5), 0, where=(x>=0) & (y5>=0), color='gray', alpha=0.3)

# Add some labels and a legend
ax.set_xlabel(r'$x_1$')
ax.set_ylabel(r'$x_2$')
ax.legend()

# Set limits
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Add grid
ax.grid(True)

# Show plot
plt.show()
```
